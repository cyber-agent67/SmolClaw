## Abstract Title

**SmolClaw: Vision-Based Autonomous Web Navigation with Entity-Interaction Architecture and A* Pathfinding**

## Abstract

Autonomous web navigation remains a critical challenge in modern software systems, particularly when websites undergo structural changes or when extracting information from dynamically-rendered content. Traditional approaches rely on hardcoded selectors or fixed DOM patterns, making them brittle and difficult to maintain. This research presents SmolClaw, a vision-based autonomous web navigator that combines vision models with intelligent pathfinding algorithms to robustly extract information from arbitrary websites without reliance on hardcoded HTML selectors.

The core contribution of this work is the implementation of a novel Entity-Interaction Model (EIM) architecture that cleanly separates state management from operational logic. The system decomposes into three primary components: entities (pure state containers), interactions (business logic and side effects), and repositories (persistence layers). This architectural approach provides significant modularity and testability advantages over traditional monolithic approaches.

SmolClaw leverages Florence vision models for semantic understanding of page content and implements advanced A* search algorithms for three critical functions: (1) finding valuable pages related to a given topic by scoring pages based on lexical relevance, (2) discovering specific elements on the current page efficiently through multiple search strategies, and (3) computing optimal navigation paths to target elements. These algorithms intelligently reduce the action-observation loop complexity that typically plagues autonomous agents.

The system incorporates BeautifulSoup-enhanced DOM parsing with semantic-only filtering, allowing the agent to focus on meaningful page elements while gracefully handling malformed HTML. The architecture supports both vision-based perception and traditional DOM tree analysis, providing robust fallback strategies when one modality fails. A persistent command-awaiting interface allows the system to function as a personal assistant capable of executing complex multi-step navigation tasks.

Experimental evaluation demonstrates that SmolClaw can successfully navigate to arbitrary pages, search for content, and extract structured information from real-world websites including repository platforms and document repositories. The system gracefully adapts to changes in website structure and provides interpretable execution traces through its agent-based reasoning framework.

This work addresses the fundamental challenge of building web agents that maintain functionality despite website evolution, contributing to more robust and maintainable autonomous systems. The modular EIM architecture provides a reusable foundation for future agent systems, while the vision-augmented pathfinding approach offers broader applicability to other spatial reasoning and navigation domains.
